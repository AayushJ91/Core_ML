```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Simple 2D dataset
X = np.array([
    [1, 2], [2, 2], [2, 3],
    [8, 7], [8, 8],
    [25, 80]
])

# Run DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=2)
labels = dbscan.fit_predict(X)

# Plot
plt.scatter(X[:,0], X[:,1], c=labels, cmap='plasma', s=100, edgecolors='black')
plt.title("DBSCAN Clustering Visualization")
plt.xlabel("X1")
plt.ylabel("X2")

# Highlight noise points
for i, txt in enumerate(labels):
    if txt == -1:
        plt.annotate("Noise", (X[i,0]+0.3, X[i,1]))

plt.show()
```
